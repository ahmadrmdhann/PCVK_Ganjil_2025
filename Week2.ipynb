## Langkah 2

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/PCVK/kitten.jpg')
plt.imshow(img)
```

## Langkah 3

Konversi warna BGR (Blue, Green, Red) menjadi RGB (Red, Green, Blue)

```python
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img3)
```

## Langkah 4

Menampilkan citra grayscale, melakukan resizing, melakukan flipping, dan meyimpan citra hasil

```python
img_gray = cv.imread('/content/drive/MyDrive/PCVK/kitten.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray, cmap='magma')
```

```python
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

```python
img_5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.imshow(img_5)
```

## Langkah 5

Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16

```python
black_img = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img, (384, 0), (510, 150), (0, 255, 0), 10)
plt.imshow(black_img)
```

7. Buat rectangle dan circle pada bagian wajah dari image foto s anda saat beraktifitas (bukan pasfoto).

```python
image = cv.imread('/content/drive/MyDrive/PCVK/Foto Aktivitas.JPG')
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Load Haar Cascade untuk deteksi wajah
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Deteksi wajah
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=8, minSize=(50, 50))

# Gambar rectangle dan circle pada wajah yang terdeteksi
for (x, y, w, h) in faces:
    # Rectangle di sekitar wajah
    cv.rectangle(image_rgb, (x, y), (x + w, y + h), (255, 0, 0), 3)  # merah

    # Circle di tengah wajah
    center = (x + w//2, y + h//2)
    radius = w//4
    cv.circle(image_rgb, center, radius, (0, 255, 0), 3)  # hijau

# Tampilkan hasil
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Rectangle dan Circle di Wajah")
plt.show()
```

9. Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran font, dan warna font yang sesuai keinginan anda.

```python
file_path = '/content/drive/MyDrive/PCVK/Foto Aktivitas.JPG'

image = cv.imread(file_path)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ambil channel B
blue_channel = image[:, :, 0]
height, width = blue_channel.shape

# Copy channel B dan buat rectangle di sudut kiri bawah
blue_copy = blue_channel.copy()
rect_start = (0, height-400)
rect_end = (450, height)
cv.rectangle(blue_copy, rect_start, rect_end, color=0, thickness=4)

# Konversi ke 3 channel agar bisa menambahkan teks berwarna
blue_rgb = cv.cvtColor(blue_copy, cv.COLOR_GRAY2BGR)

# Tambahkan teks nama file
text_pos = (10, height-20)
cv.putText(blue_rgb, "ramadhan.jpg",
            text_pos,
            fontFace=cv.FONT_HERSHEY_SIMPLEX,
            fontScale=3,
            color=(0, 0, 0),
            thickness=3,
            lineType=cv.LINE_AA)

# Tampilkan hasil
plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
plt.imshow(blue_rgb)
plt.axis('off')
plt.show()
```
